```python
def convert(name):
    '''Converting column names to lower case'''
    pattern = re.compile(r"[A-Z]{2,}(?=[A-Z][a-z]+[0-9]*|\b)|[A-Z]?[a-z]+[0-9]*|[A-Z]|[0-9]+")
    found = pattern.findall(name)
    s1 = "_".join(found)
    return s1.lower()
```
