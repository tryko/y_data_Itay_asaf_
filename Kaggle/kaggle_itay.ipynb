```python
def convert(name):
    '''Converting column names to lower case'''
    pattern = re.compile(r"[A-Z]{2,}(?=[A-Z][a-z]+[0-9]*|\b)|[A-Z]?[a-z]+[0-9]*|[A-Z]|[0-9]+")
    found = pattern.findall(name)
    s1 = "_".join(found)
    return s1.lower()
```

```python
def rf_fit(n_estimators=100,rows = X_train.index, cols=X_train.columns, test_rows=X_test.index, title=''):
    rf = RandomForestRegressor(n_estimators=n_estimators)
    rf.fit(X_train.loc[rows, cols], y_train.loc[rows])

    y_pred = rf.predict(X_test.loc[test_rows, cols])
    min_val = y_test.loc[test_rows].min()
    max_val = y_test.loc[test_rows].max()
    print(title)
    print(mean_squared_error(y_test.loc[test_rows], y_pred, squared=False))
    print('Normalized')
    score = mean_squared_error(y_test.loc[test_rows], y_pred, squared=False)/ (max_val - min_val)
    print(score)
    print()

    
```

Next: lets see the mistakes: where do they all miss-classified and where all correctly classified.

```python
res_clfs = [(y_pred_non_numeric, y_test_non_numeric), (y_pred_numeric, y_test_numeric), (y_pred_numeric_and_non_numeric, y_test_numeric_and_non_numeric)]
def get_miss_correct_idx_by_class(res_clfs, y_price_cat):
    # y_true_type_idx = [y_price_cat[y_price_cat == i].index for i in range(4)]
    price_cat_3 = {'true_miss': np.array([]), 'pred_miss': np.array([]), 'correct': np.array([])}
    price_cat_2 = {'true_miss': np.array([]), 'pred_miss': np.array([]), 'correct': np.array([])}
    price_cat_1 = {'true_miss': np.array([]), 'pred_miss': np.array([]), 'correct': np.array([])}
    price_cat_0 = {'true_miss': np.array([]), 'pred_miss': np.array([]), 'correct': np.array([])}
    price_cats = [ price_cat_0, price_cat_1, price_cat_2, price_cat_3, ]

    for res in res_clfs:
        for price_class in range(4):

            true_res_idx = np.where(res[1] == price_class)
            pred_res_idx = np.where(res[0] == price_class)
            
            # Classified class incorrectly
            true_miss_idx = np.setdiff1d( true_res_idx, pred_res_idx)
            # Did not classify class correctly
            pred_miss_idx = np.setdiff1d( pred_res_idx, true_res_idx)
            
            correct_idx = np.union1d(true_res_idx, pred_res_idx)
            price_cats[price_class]['true_miss'] = np.append(price_cats[price_class]['true_miss'], true_miss_idx)
            price_cats[price_class]['pred_miss'] = np.append(price_cats[price_class]['pred_miss'] ,pred_miss_idx)
            price_cats[price_class]['correct'] = np.append(price_cats[price_class]['correct'], correct_idx)
    return price_cats


miss_or_correct = get_miss_correct_idx_by_class(res_clfs=res_clfs, y_price_cat=y_price_cat)



for i in range(4):
    true_miss_clf[i] = np.append(true_miss_clf[i],miss_or_correct[i]['true_miss'])
    pred_miss_clf[i] = np.append(pred_miss_clf[i], miss_or_correct[i]['pred_miss'])
    correct_clf[i] = np.append(correct_clf[i], miss_or_correct[i]['correct'])

```

```python
def get_miss_correct():
    true_miss_clf = [np.array([]), np.array([]), np.array([]), np.array([])]
    pred_miss_clf = [np.array([]), np.array([]), np.array([]), np.array([])]
    correct_clf = [np.array([]), np.array([]), np.array([]), np.array([])]

    for _ in range(100):
        RANDOM_STATE = np.random.random_integers(low=0, high=255, size=1)[0]

        high_price = y[y > 214000].index
        medium_price = y[(y <= 214000) & (y > 162000)].index
        medium_low_price = y[(y <= 162000) & (y > 129500)].index
        low_price = y[(y <= 129500) ].index


        price_categories = [high_price, medium_price, medium_low_price, low_price]
        idx = np.concatenate(price_categories)

        y_price_cat = y.loc[idx].copy()
        y_price_cat.loc[high_price] = 3
        y_price_cat.loc[medium_price] = 2
        y_price_cat.loc[medium_low_price] = 1
        y_price_cat.loc[low_price] = 0
        
        X_non_numeric = pd.get_dummies(X[non_numeric_cols], dummy_na=True) # gives better score than with out na
        X_non_numeric = X_non_numeric.loc[idx].copy()

        X_train_non_numeric, X_test_non_numeric, y_train_non_numeric, y_test_non_numeric = train_test_split(X_non_numeric, y_price_cat, test_size=0.2, random_state=RANDOM_STATE)

        ## Classification on non_numeric
        rf_clf = RandomForestClassifier(n_estimators=100)
        rf_clf.fit(X_train_non_numeric, y_train_non_numeric)
        y_pred_non_numeric = rf_clf.predict(X_test_non_numeric)
        y_pred_non_numeric = pd.Series(y_pred_non_numeric, index=y_test_non_numeric.index)
        (y_pred_non_numeric == y_test_non_numeric).mean()


        ## Classification on numeric:
        X_numeric = X.loc[:, numeric_no_na_cols]
        X_numeric = X_numeric.loc[idx].copy()

        X_train_numeric, X_test_numeric, y_train_numeric, y_test_numeric = train_test_split(X_numeric, y_price_cat, test_size=0.2, random_state=RANDOM_STATE)

        rf_clf = RandomForestClassifier(n_estimators=100)
        rf_clf.fit(X_train_numeric, y_train_numeric)
        y_pred_numeric = rf_clf.predict(X_test_numeric)

        y_pred_numeric = pd.Series(y_pred_numeric, index=X_test_numeric.index)
        (y_pred_numeric == y_test_numeric).mean()

        ## Classification based on both
        X_numeric_and_non_numeric = pd.merge(X_numeric,X_non_numeric , left_index=True, right_index=True)

        X_numeric_and_non_numeric = X_numeric_and_non_numeric.loc[idx].copy()

        X_train_numeric_and_non_numeric, X_test_numeric_and_non_numeric, y_train_numeric_and_non_numeric, y_test_numeric_and_non_numeric = train_test_split(X_numeric_and_non_numeric, y_price_cat, test_size=0.2, random_state=RANDOM_STATE)

        rf_clf = RandomForestClassifier(n_estimators=100)
        rf_clf.fit(X_train_numeric_and_non_numeric, y_train_numeric_and_non_numeric)
        y_pred_numeric_and_non_numeric = rf_clf.predict(X_test_numeric_and_non_numeric)

        y_pred_numeric_and_non_numeric = pd.Series(y_pred_numeric_and_non_numeric, index=X_test_numeric_and_non_numeric.index)
        (y_pred_numeric_and_non_numeric == y_test_numeric_and_non_numeric).mean()

        res_clfs = [(y_pred_non_numeric, y_test_non_numeric), (y_pred_numeric, y_test_numeric), (y_pred_numeric_and_non_numeric, y_test_numeric_and_non_numeric)]
        def get_miss_correct_idx_by_class(res_clfs, y_price_cat):
            # y_true_type_idx = [y_price_cat[y_price_cat == i].index for i in range(4)]
            price_cat_3 = {'true_miss': np.array([]), 'pred_miss': np.array([]), 'correct': np.array([])}
            price_cat_2 = {'true_miss': np.array([]), 'pred_miss': np.array([]), 'correct': np.array([])}
            price_cat_1 = {'true_miss': np.array([]), 'pred_miss': np.array([]), 'correct': np.array([])}
            price_cat_0 = {'true_miss': np.array([]), 'pred_miss': np.array([]), 'correct': np.array([])}
            price_cats = [ price_cat_0, price_cat_1, price_cat_2, price_cat_3, ]

            for res in res_clfs:
                for price_class in range(4):

                    true_res_idx = np.where(res[1] == price_class)
                    pred_res_idx = np.where(res[0] == price_class)
                    
                    # Classified class incorrectly
                    true_miss_idx = np.setdiff1d( true_res_idx, pred_res_idx)
                    # Did not classify class correctly
                    pred_miss_idx = np.setdiff1d( pred_res_idx, true_res_idx)
                    
                    correct_idx = np.union1d(true_res_idx, pred_res_idx)
                    price_cats[price_class]['true_miss'] = np.append(price_cats[price_class]['true_miss'], true_miss_idx)
                    price_cats[price_class]['pred_miss'] = np.append(price_cats[price_class]['pred_miss'] ,pred_miss_idx)
                    price_cats[price_class]['correct'] = np.append(price_cats[price_class]['correct'], correct_idx)
            return price_cats


        miss_or_correct = get_miss_correct_idx_by_class(res_clfs=res_clfs, y_price_cat=y_price_cat)



        for j in range(4):
            true_miss_clf[j] = np.append(true_miss_clf[j],miss_or_correct[j]['true_miss'])
            pred_miss_clf[j] = np.append(pred_miss_clf[j], miss_or_correct[j]['pred_miss'])
            correct_clf[j] = np.append(correct_clf[j], miss_or_correct[j]['correct'])

    true_miss_v_cnt = [np.array([]), np.array([]), np.array([]), np.array([])]
    pred_miss_v_cnt = [np.array([]), np.array([]), np.array([]), np.array([])]
    correct_v_cnt = [np.array([]), np.array([]), np.array([]), np.array([])]

    for i in range(4):
        true_miss_v_cnt[i] = pd.value_counts(true_miss_clf[i])
        pred_miss_v_cnt[i] = pd.value_counts(pred_miss_clf[i])
        correct_v_cnt[i] = pd.value_counts(correct_clf[i])

    return true_miss_v_cnt, pred_miss_v_cnt, correct_v_cnt
```
