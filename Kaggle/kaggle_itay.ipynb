```python
def convert(name):
    '''Converting column names to lower case'''
    pattern = re.compile(r"[A-Z]{2,}(?=[A-Z][a-z]+[0-9]*|\b)|[A-Z]?[a-z]+[0-9]*|[A-Z]|[0-9]+")
    found = pattern.findall(name)
    s1 = "_".join(found)
    return s1.lower()
```

```python
def rf_fit(n_estimators=100,rows = X_train.index, cols=X_train.columns, test_rows=X_test.index, title=''):
    rf = RandomForestRegressor(n_estimators=n_estimators)
    rf.fit(X_train.loc[rows, cols], y_train.loc[rows])

    y_pred = rf.predict(X_test.loc[test_rows, cols])
    min_val = y_test.loc[test_rows].min()
    max_val = y_test.loc[test_rows].max()
    print(title)
    print(mean_squared_error(y_test.loc[test_rows], y_pred, squared=False))
    print('Normalized')
    score = mean_squared_error(y_test.loc[test_rows], y_pred, squared=False)/ (max_val - min_val)
    print(score)
    print()

    
```
