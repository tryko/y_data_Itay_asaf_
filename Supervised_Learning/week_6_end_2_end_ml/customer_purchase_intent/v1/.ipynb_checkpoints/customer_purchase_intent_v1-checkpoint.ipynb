### 2.1 Evaluation pipeline
bulding a sliding window evaluation pipeline
Models configurations
Results analysis

```python
from dateutil.relativedelta import relativedelta
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import precision_recall_curve

def evaluate(events_df, models,feature_extractors_conf, days_train, days_val, days_test, days_slide,
            limit_folds = 0):
    print("creating the cross validation dates")
    cv = []
    train_start = events_df['date'].min()
    while True:
        train_end = train_start + relativedelta(days=days_train)
        validation_end = train_end+ relativedelta(days=days_val)
        test_end = validation_end+ relativedelta(days=days_test)
        if test_end < events_df['date'].max():
            cv.append(
                dict(train_start = train_start, train_end = train_end,validation_end = validation_end,
                    test_end = test_end))
            train_start = train_start + relativedelta(days=days_slide)
        else:
            break

    results = []
    batch_datasets = {}
    batch_models = {}
    if limit_folds > 0:
        cv = cv[:limit_folds]
    for cv_ in cv:
        batch = cv_['validation_end']
        print(f'batch test start at {cv_["validation_end"]}')
        train_df = events_df[(events_df['date'] > cv_['train_start']) & (events_df['date'] <= cv_['train_end'])]
        validation_df = events_df[(events_df['date'] > cv_['train_end']) & (events_df['date'] <= cv_['validation_end'])]
        test_df = events_df[(events_df['date'] > cv_['validation_end']) & (events_df['date'] <= cv_['test_end'])]
        
        for feature_extractor_name, feature_extractor_conf in feature_extractors_conf.items():
            print('prep data')
            feature_extractor = feature_extractor_conf["class"](**feature_extractor_conf["args"])
            train_X, train_y = feature_extractor.prep_raw(train_df)
            val_X, val_y = feature_extractor.prep_raw(validation_df)
            test_X, test_y = feature_extractor.prep_raw(test_df)

            print('extract features')
            feature_extractor.fit(train_X,train_y)
            train_features = feature_extractor.transform(train_X)
            val_features = feature_extractor.transform(val_X)
            test_features = feature_extractor.transform(test_X)

            X_for_grid_cv = pd.concat([train_features,val_features]).reset_index(drop = True)
            y_for_grid_cv = pd.concat([train_y,val_y]).reset_index(drop = True)
            cv_for_grid_cv = [[list(range(0,train_y.shape[0])), list(range(train_y.shape[0],y_for_grid_cv.shape[0]))]]
            batch_datasets[str(batch)] = dict(test_df = test_df, test_X = test_X, test_features = test_features, test_y=test_y)
            batch_models[str(batch)] = dict()
            for model_name, model_params in models.items():
                print(f'fit {model_name}')
                clf_cv = GridSearchCV(model_params['class'](),model_params['args'], cv = cv_for_grid_cv)

                clf_cv.fit(X_for_grid_cv,y_for_grid_cv)
                preds_test = clf_cv.predict(test_features)
                probs_test = clf_cv.predict_proba(test_features)[:,1]
                
                preds_val = clf_cv.predict(val_features)
                probs_val = clf_cv.predict_proba(val_features)[:,1]
              
                precision, recall, thresholds =  precision_recall_curve(val_y,probs_val)
                optimized_threshold = thresholds[(2 * (precision * recall) / (precision + recall)).argmax()]
                preds_optimized_test = [probs_test > optimized_threshold][0].astype(int)

                model_res = dict(batch = str(batch), model_name=model_name, best_params = clf_cv.best_params_, 
                                 features = feature_extractor_name,
                                 accuracy_score = accuracy_score(test_y, preds_test),
                                 f1_score = f1_score(test_y, preds_test),
                                 precision = precision_score(test_y, preds_test),
                                 recall = recall_score(test_y, preds_test),
                                 f1_score_optimized=f1_score(test_y, preds_optimized_test),
                                 optimized_threshold = optimized_threshold)
                batch_models[str(batch)][model_name] = clf_cv
                results.append(model_res)
    return results, batch_datasets, batch_models
```

```python
def analyze_results(results):
    pd.DataFrame(results)
    pd.DataFrame(results).groupby(['features','model_name'])[['f1_score_optimized','f1_score']].mean()
    results_ = pd.DataFrame(results).groupby(['batch','model_name'])['f1_score_optimized'].mean().reset_index()
    results_['batch']= pd.to_datetime(results_['batch'])
    plt.figure()
    sns.set(rc={'figure.figsize':(100,30)})
    sns.set(font_scale=10)
    sns.lineplot(data=results_, x="batch", y="f1_score_optimized", hue="model_name")
```

### 2.2 Feature Extraction classes
wrap features into a class

```python
from sklearn.preprocessing import StandardScaler
class FeatureExtractor(object):
    def __init__(self, session_size = 3):
        self.session_size = session_size
        self.apply_list = ['category_id','brand','category_code','product_id']
        self.session_keys = ['user_session','user_id']
        
    def prep_raw(self, df) -> (pd.DataFrame, pd.Series):
        pass
        
    def fit(self, X, y = None):
        pass
    
    def fit_transform(self, X, y = None) -> (pd.DataFrame, pd.Series):
        pass
    
    def transform(self, X, y = None) -> pd.DataFrame:
        pass
    
class FeatureExtractorV1(FeatureExtractor):
    def __init__(self, session_size = 3):
        super().__init__(session_size)
        
        self.features_aggs = {'user_session' : ['count'],
                              'weekday' : ['first'],
                              'hour' : ['first'],                        
                                'price':['mean','std'],
                               'category_code':['nunique','count'],
                              'category_id':['nunique','count'],
                               'product_id':['nunique'],
                               'brand':['nunique','count']
                } 
        self.features_types = {
                          'price_mean': 'float',
                          'price_std' : 'float',
                           'category_id': 'list',
                           'brand' : 'list',
                           'product_id' : 'list',
                          'weekday_first' : 'categorical',
                          'hour_first': 'categorical'}
        
        self.scaler = StandardScaler()
        
        
    def prep_raw(self, df):
        first_sessions = df.sort_values('datetime').groupby(self.session_keys).head(3)
        
        first_sessions_features = first_sessions.groupby(self.session_keys).agg(self.features_aggs)
        first_sessions_features.columns = ['_'.join(col).strip() for col in first_sessions_features.columns.values]
        
        for c in self.apply_list:
            first_sessions_features[c] = first_sessions.groupby(self.session_keys)[c].apply(list)
        
        first_sessions_features['label'] = first_sessions.groupby(self.session_keys)['event_type'].apply(lambda x: x.eq('cart').any() | x.eq('purchase').any()).astype(int)
        first_sessions_features = first_sessions_features[first_sessions_features['user_session_count']==3]
        X = first_sessions_features.drop('label', axis=1)
        y = first_sessions_features['label']
        return X, y
    
    def get_features_by_type(self,f_type):
        return [f for f, t in self.features_types.items() if t == f_type]
        
    def fit(self, X, y = None):
        features = self.transform(X)
        self.scaler.fit(features[self.get_features_by_type('float')])
    
    def transform(self, X, y=None):
        features = X.copy()
        float_f = self.get_features_by_type('float')
        try:
            features[float_f] = self.scaler.transform(features[float_f])
        except:
            print("error in transform scalar, continue without scaling")
        features[float_f] = features[float_f].fillna(0)
        features = features.drop(self.apply_list, axis = 1)
        return features
    

```

### Adding new features into the configuration - In class

```python
class FeatureExtractorV2(FeatureExtractor):
    def __init__(self, session_size = 3):
        super().__init__(session_size)
        self.base_features = FeatureExtractorV1(session_size)
        self.prior_ratio = {}
        
    def prep_raw(self, df):
        return self.base_features.prep_raw(df)
        
    def fit(self, X, y = None):
        for agg_key in self.base_features.apply_list:
            view_full = X[X['event_type'] =='view'].groupby(agg_key)['user_session'].count()
            cart_full = X[X['event_type'] =='cart'].groupby(agg_key)['user_session'].count()
            cart_ratio = cart_full/ view_full
            cart_ratio = cart_ratio.fillna(0)
            self.prior_ratio[agg_key] = cart_ratio.to_dict()
        
        self.base_features.fit(X, y)
    
    def transform(self, X, y=None):
        base_features = self.base_features.transform(X)
        self_features = self.transform_(X)
        features = pd.concat([base_features,self_features], axis = 1)
        return features
    
    def transform_(self, X, y=None):
        features = X.copy()
        #TODO Add your features here
        return features
```

```python
def get_data(csv_path, file_name):
    raw_prep_path = os.path.join(csv_path, f"{file_name}_prep1.csv")
    if not os.path.isfile(raw_prep_path):
        events_df = pd.read_csv(os.path.join(csv_path, f"{file_name}.csv"))
        events_df['datetime'] = events_df['event_time'].apply(lambda x : pd.to_datetime(x,infer_datetime_format=True))
        events_df['date'] = events_df['datetime'].apply(lambda x : x.date())
        events_df['weekday'] = events_df['datetime'].apply(lambda x: x.weekday)
        events_df['hour'] = events_df['datetime'].apply(lambda x: x.hour)
        events_df.to_csv(raw_prep_path,index = False)
    events_df= pd.read_csv(raw_prep_path, parse_dates=['datetime','date'])
    return events_df
```
